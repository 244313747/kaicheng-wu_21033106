```python
#Question 3
#data from https://github.com/cmoa/collection
with open('data/cmoa_clean.tsv', encoding='utf-8') as f:
    cmoa_data = np.loadtxt(f, dtype = 'O', skiprows=1, delimiter="\t") 
    cmoa_data[:,3] = cmoa_data[:,3].astype(float)
    cmoa_data[:,4] = cmoa_data[:,4].astype(float)
    cmoa_data = np.array(cmoa_data)
```

```python
import collections

# Get the artist names from the data array
artists = cmoa_data[:,5]

# Create a Counter object to keep track of the group sizes
group_sizes = collections.Counter()

# Loop through each row in the data array
for i in artists:
    # Split the artist name at the "|" character
    group_size = len(str(i).split('|'))
    # Increment the count for the group size in the Counter object
    group_sizes[group_size] += 1

# Print the group sizes and their counts
for group_size, count in group_sizes.items():
    print(f'{group_size} artists: {count}')
```

```python
#Question 3b
#data from #https://corgis-edu.github.io/corgis/csv/food/
with open('data/unique_foods.csv') as f:
    food = np.loadtxt(f, dtype = 'O', skiprows=1, delimiter="\t") 
    food = np.array(food)
```

```python
# Create a counter variable to keep track of the number of artworks that contain a food from the list
counter = 0

# Loop through each row in the data array
for row in cmoa_data:
    # Get the title of the artwork
    title = row[1]
    # Loop through each element of the food list
    for food_list in food:
        # Loop through each food in the food list
        for f in food_list:
            # Check if the food is a whole word in the title
            if any(word == f for word in title.split()):
                # If the food is in the title, increment the counter variable
                counter += 1
                # Stop looping through the food list
                break


# Calculate the percentage of artworks that contain a food from the list
percentage = counter / len(cmoa_data)*100

# Print the percentage of artworks that contain a food from the list
print(percentage + "%")
```
